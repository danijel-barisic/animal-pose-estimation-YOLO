### Train, Val and Test preparation

```python
import os
import shutil


labels_dir = "D:\\Tomi\\Faks\\neumreYolo\\labels\\ap10k-train-split1"
source_dir = "D:\\Tomi\\Faks\\neumreYolo\\ap-10k\\data"

target_dir = "D:\\Tomi\\Faks\\neumreYolo\\dataset\\train\\images"

os.makedirs(target_dir, exist_ok=True)

label_files = [f for f in os.listdir(labels_dir) if f.endswith('.txt')]

for label_file in label_files:
    base_name = os.path.splitext(label_file)[0]
    source_image_path = os.path.join(source_dir, base_name + '.jpg')
    print(source_image_path)
    if os.path.exists(source_image_path):
        target_image_path = os.path.join(target_dir, base_name + '.jpg')
        shutil.copy(source_image_path, target_image_path)
        print(f"copied: {source_image_path} -> {target_image_path}")
    else:
        print(f"No corresponding image found for {label_file}")
```
